```python
import math
import matplotlib.pyplot as plt
```

```python
# Get input from the user
numbers = []
for i in range(4):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)
```

```python
def find_log10_values(numbers):
    """
    Calculates the base-10 logarithm of a list of numbers.

    Args:
      numbers: A list of numbers.

    Returns:
      A list of the base-10 logarithms of the input numbers.
    """
    log10_values = []
    for num in numbers:
        if num <= 0:
            log10_values.append("Undefined")  # Logarithm of non-positive numbers is undefined
        else:
            log10_values.append(math.log10(num))  # Use math.log10 for base-10 logarithm
    return log10_values

# Calculate log10 values
log10_values = find_log10_values(numbers)
```

```python
# Create lists for plotting
x_values = list(range(1, len(numbers) + 1))  # Use indices for x-axis
y_values = numbers
y_log10_values = [value for value in log10_values if value != "Undefined"]  # Filter out undefined values

# Plot the numbers (first graph)
plt.figure(1)  # Create a new figure for the first graph
plt.plot(x_values, y_values, color='red', marker='o', label='Numbers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of Numbers')
plt.legend()
plt.grid(True)
plt.savefig('numbers_plot.png')  # Save the first graph

# Plot the log10 values (second graph)
plt.figure(2)  # Create a new figure for the second graph
plt.plot(x_values[:len(y_log10_values)], y_log10_values, color='blue', marker='s', label='Log10 Values')
plt.xlabel('Index')
plt.ylabel('Log10 Value')
plt.title('Plot of Log10 Values')
plt.legend()
plt.grid(True)
plt.savefig('log10_values_plot.png')  # Save the second graph

# Show the plots
plt.show()
```
